```python
#RGB 3-tuple in python, adding RGB value for each color. 
colors = [
    [
        [0,0,255], #blue
        [0,255,0] #green
    ],
    [
        [255,0,0], #red
        [255,255,0] #yellow
    ]
]

print(np.array(colors).shape)

plt.imshow(colors)
start_row = 0
for row in colors:
    start_col = -0.25
    for color in row:
        plt.text(start_col, start_row, str(color))
        start_col += 1
    start_row += 1
```

```python
#read image using OpenCV
image = cv2.imread("./data/Venice_2013-69.jpg")
```

```python
#image type
type(image)
```

```python
#image displayed as arrays in numpy
image.shape
```

```python
plt.imshow(image)
```

```python
# parse BGR to RGB (color format)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# display image in color format
plt.imshow(image)
```

```python
# parse image to grayscale
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# display image
plt.imshow(gray_img, cmap='gray')
```

```python
#resizing image and ignoring the aspect ratio
WIDTH = 300
HEIGHT = 300
```

```python
resized = cv2.resize(image, (WIDTH, HEIGHT))
plt.imshow(resized)
```

```python
#rescaling for landscape vs portrait images
aspect = image.shape[1] / float(image.shape[0])
print("Image aspect: %f" % aspect)

if(aspect > 1):
    #landscape orientation
    res = int(aspect * HEIGHT)
    scaled = cv2.resize(image, (res, HEIGHT))
if(aspect < 1):
    #portrait orientation
    res = int(WIDTH / aspect)
    scaled = cv2.resize(image, (WIDTH, res))
if(aspect == 1):
    scaled = cv2.resize(image, (WIDTH, HEIGHT))
    
plt.imshow(scaled)
```

```python
# Cropping the image in the center
def crop_center(img, cropx, cropy):
    y,x,c = img.shape
    startx = x//2-(cropx//2)
    starty = y//2-(cropy//2)
    return img[starty:starty+cropy, startx:startx+cropx]

#scaled image
cropped = crop_center(scaled, WIDTH, WIDTH)

plt.imshow(cropped)
```

```python
image = cropped.copy()

#adding text to the image
cv2.putText(image, "Burano", (100,270),
            cv2.FONT_HERSHEY_COMPLEX, 0.8, (20, 240, 150), 2)

# add line
cv2.line(image, (100,280), (200,280), (50, 100, 250), 3)

# show image
plt.imshow(image)
```
